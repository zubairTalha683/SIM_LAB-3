```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns



random_numbers = np.random.rand(10)


plt.hist(random_numbers, bins=50,  alpha=0.5, color='yellow', edgecolor='black')


plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)


plt.show()
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns


NormalValues = np.random.normal(0, 1, 500)


plt.hist(NormalValues, bins=30, edgecolor='red')
plt.title('Histogram of Normal Distribution (mean=100, std=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# With Seed
np.random.seed(4)  # Setting seed to 10
seeded_matrix = np.random.randint(1, 50, (3, 3))
print("Seeded Random Matrix:\n", seeded_matrix)

```

```python
# Without Seed
np.random.seed(None)  # No seed, generates different values each time
non_seeded_matrix = np.random.randint(1, 50, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)
```
